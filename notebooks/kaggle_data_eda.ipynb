# Kaggle data

The data for OCT project is hosted both on [Kaggle](https://www.kaggle.com/datasets/paultimothymooney/kermany2018) and [Mendeley](https://data.mendeley.com/datasets/rscbjbr9sj/2). The latter is referenced by the writers of the paper [here](https://www.cell.com/cell/fulltext/S0092-8674(18)30154-5). 

We are first going to check if these 2 datasets are the same. 
After running the following code we reazlize that the data on Kaggle has been copied twice and therefore it is twice as big. This notebook is to show why we are going to use the dataset on Mendeley. 

EDA is done in **data-ingestion** notebook on Mendeley dataset. 

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import numpy as np
```

```python
# removed the whitespace from OCT2017
BASE_DATA_PATH = '../data/processed/oct2017/OCT2017'
TRAIN_BASE_PATH = '../data/processed/oct2017/OCT2017/train/'
TEST_BASE_PATH = '../data/processed/oct2017/OCT2017/test/'
VAL_BASE_PATH = '../data/processed/oct2017/OCT2017/val'
```

```python
parents = ['train', 'test', 'val']
children = ['CNV', 'DME', 'DRUSEN', 'NORMAL']

# create a base dataframe from the directory items
names_cache = []
for p in parents:
    for c in children:
        files = os.listdir(os.path.join(BASE_DATA_PATH, p, c))
        instance_type = [ p for i in range(len(files)) ]
        path = [ f'/OCT2017/{p}/{c}' for i in range(len(files))]
        print(f'{p}/{c} contains {len(files)} files')
        names_cache.extend(list(zip(files, path, instance_type)))

oct_df = pd.DataFrame(names_cache, columns=['filename', 'path', 'type'])

#remove the .DS_store filenames
oct_df = oct_df[~oct_df['filename'].str.contains('.DS_Store')]
oct_df.reset_index(drop=True,inplace=True)

# split filename into features
oct_df['disease'] = oct_df['filename'].apply(lambda f: f.split('-')[0])
oct_df['pat_id'] = oct_df['filename'].apply(lambda f: f.split('-')[1])
oct_df['img_id'] = oct_df['filename'].apply(lambda f: f.split('-')[2].split('.')[0])
oct_df['filetype'] = oct_df['filename'].apply(lambda f: f.split('.')[1])
oct_df.head()
```
